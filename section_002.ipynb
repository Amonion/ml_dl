```python
import numpy as np  # For numerical computing and working with arrays
import pandas as pd  # For handling and analyzing data
```

```python
# Integer and float
x = 10
y = 3.5

# String
name = "Machine Learning"

# List
numbers = [1, 2, 3, 4, 5]

# Dictionary
student = {"name": "John", "age": 20, "course": "ML"}

print(x, y, name)
print(numbers)
print(student["name"])  # Access dictionary value

# Loop through a list
fruits = ["Apple", "Banana", "Berry", "Cherry"]
for fruit in fruits:
    print(f"I Love this fruit called {fruit}")
```

```python
# Print numbers from 1 to 5
num = 1
while num <= 5:
    print(num)
    num += 1  # Increment num

def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

```python
#Creating a 1D array
arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:", arr1)

# Creating a 2D array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2)

#Array Addition
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
sum_array = a + b
print("Sum:", sum_array)

# Mean and Standard Deviation
print("Mean:", np.mean(a))
print("Standard Deviation:", np.std(a))

# NumPy performs fast mathematical operations on arrays.

# np.mean() calculates the average, while np.std() finds data variation.
```

```python
# Creating a simple dataset
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Score": [85, 90, 95]
}

df = pd.DataFrame(data)
print(df)
# pandas.DataFrame structures data like an Excel table for analysis.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='--', color='r', label="Sales")
plt.xlabel("Days")
plt.ylabel("Revenue")
plt.title("Sales Over Time")
plt.legend()
plt.show()

# Matplotlib is used to create charts.

# Line plots show trends over time.
```

```python
# Generate random data
data = np.random.randn(1000)

# Create histogram
sns.histplot(data, bins=30, kde=True)
plt.title("Distribution of Random Data")
plt.show()
```

```python
# Sample Data
df = pd.DataFrame({"x": np.random.rand(50), "y": np.random.rand(50)})

sns.scatterplot(x="x", y="y", data=df)
plt.title("Random Scatter Plot")
plt.show()

### Histograms show data distribution.
# In the histogram, bins=30 determines the number of bars (bins) used to divide the data range.

# How It Works:
# A histogram groups continuous data into intervals (bins) and counts the occurrences in each bin.

# bins=30 means the data range will be divided into 30 equal-width intervals.

# More bins → Finer granularity (more details).

# Fewer bins → More general overview (less detail).

###
# kde=True adds a smooth curve to the histogram.
```
